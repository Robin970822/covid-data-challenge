```python
import os
import pandas as pd

def get_Prognosis(prediction, thres=0.5):
    return 'SEVERE' if prediction > thres else 'MILD'
```

```python
# method = 'knn_9'
method = 'softimpute'
```

```python
prediction_path = './csv/prediction_testSet/filled_{}_binary.txt'.format(method)
```

```python
prediction_result = pd.read_csv(prediction_path, index_col=0)
prediction_result['Prognosis'] = prediction_result['prediction_clinical'].map(get_Prognosis)
```
